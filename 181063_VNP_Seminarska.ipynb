**Припремање на податоците**

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

df['satisfaction_level'] = label_encoder.fit_transform(df['satisfaction_level'])
df['Departments'] = label_encoder.fit_transform(df['Departments'])
df['salary'] = label_encoder.fit_transform(df['salary'])
```

**Градење, тренирање и тестирање на моделот**

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
X = df.drop(['satisfaction_level'], axis=1)
y = df['satisfaction_level']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
rfc = RandomForestClassifier()

rfc.fit(X_train, y_train)

y_predict = rfc.predict(X_test)

accuracy = accuracy_score(y_test, y_predict)
precision = precision_score(y_test, y_predict, average='macro')
recall = recall_score(y_test, y_predict, average='macro')
f1 = f1_score(y_test, y_predict, average='macro')

print('Accuracy:', accuracy)
print('Precision:', precision)
print('Recall:', recall)
print('F1-score:', f1)
```

**Оптимизирање на хиперпараметрите**

```python
from sklearn.model_selection import validation_curve

n_estimators_range = [100, 200, 300, 400]

valcurve = validation_curve(RandomForestClassifier(), X_train, y_train,
                           param_name='n_estimators', param_range=n_estimators_range, cv=5)

train_score = valcurve[0].mean(axis=1)
test_score = valcurve[1].mean(axis=1)

d1 = pd.DataFrame({'Training Set': train_score, 'Testing Set': test_score}, index=n_estimators_range)

ax = d1.plot(figsize=(10, 6), linestyle='-', marker='o')
ax.set_title('Validation Curve for n_estimators')
ax.set_xlabel('n_estimators')
ax.set_ylabel('Cross Validation Score')
ax.legend(loc='best')
plt.show()
```

```python
max_depth_range = [5, 10, 15, 20, 25]

valcurve = validation_curve(RandomForestClassifier(n_estimators=300), X_train, y_train,
                            param_name='max_depth', param_range=max_depth_range, cv=5)

train_score = valcurve[0].mean(axis=1)
test_score = valcurve[1].mean(axis=1)

df_scores = pd.DataFrame({'Training Set': train_score, 'Testing Set': test_score}, index=max_depth_range)

ax = df_scores.plot(figsize=(10, 6), linestyle='-', marker='o')
ax.set_title('Validation Curve for max_depth')
ax.set_xlabel('max_depth')
ax.set_ylabel('Cross Validation Score')
ax.legend(loc='best')
plt.show()
```

```python
min_samples_split_range = [2, 5, 10, 15, 20]

valcurve = validation_curve(RandomForestClassifier(n_estimators=300, max_depth=20), X_train, y_train,
                            param_name='min_samples_split', param_range=min_samples_split_range, cv=5)

train_score = valcurve[0].mean(axis=1)
test_score = valcurve[1].mean(axis=1)

df_scores = pd.DataFrame({'Training Set': train_score, 'Testing Set': test_score}, index=min_samples_split_range)

ax = df_scores.plot(figsize=(10, 6), linestyle='-', marker='o')
ax.set_title('Validation Curve for min_samples_split')
ax.set_xlabel('min_samples_split')
ax.set_ylabel('Cross Validation Score')
ax.legend(loc='best')
plt.show()
```

```python
min_samples_leaf_range = [1, 2, 4, 8, 12]

valcurve = validation_curve(RandomForestClassifier(n_estimators=300, max_depth=20), X_train, y_train,
                            param_name='min_samples_leaf', param_range=min_samples_leaf_range, cv=5)

train_score = valcurve[0].mean(axis=1)
test_score = valcurve[1].mean(axis=1)

df_scores = pd.DataFrame({'Training Set': train_score, 'Testing Set': test_score}, index=min_samples_leaf_range)

ax = df_scores.plot(figsize=(10, 6), linestyle='-', marker='o')
ax.set_title('Validation Curve for min_samples_leaf')
ax.set_xlabel('min_samples_leaf')
ax.set_ylabel('Cross Validation Score')
ax.legend(loc='best')
plt.show()
```

```python
rfc = RandomForestClassifier(n_estimators=300, max_depth=20)
model = rfc.fit(X_train, y_train)
y_predict = model.predict(X_test)

accuracy = accuracy_score(y_test, y_predict)
precision = precision_score(y_test, y_predict, average='macro')
recall = recall_score(y_test, y_predict, average='macro')
f1 = f1_score(y_test, y_predict, average='macro')

print('Accuracy:', accuracy)
print('Precision:', precision)
print('Recall:', recall)
print('F1-score:', f1)
```
